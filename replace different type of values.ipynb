```python
# importamos la libreria pandas
import pandas as pd
# importamos la libreria numpy
import numpy as np
```

```python
# creamos un dataframe para el ejemplo
df = pd.DataFrame({
    "codigo": [1,2,3,4],
    "ciudad": ["santiago","concepcion","temuco","Punta arenas"],
    "cantidad_habitantes": [300,200,None,400]
})

df
```

```python
# generamos una copia del data frame original
df1 = df.copy()
df1
```

```python
# descripcion estad√≠stica de la columna 'cantidad_habitantes' 
df['cantidad_habitantes'].describe()
```

```python
# almacenamos la media aritmetica de la columna 'cantidad_habitantes' 
mean_habitantes = df['cantidad_habitantes'].mean()
mean_habitantes
```

```python
# reemplazamos los datos nulos de la columna 'cantidad_habitantes' 
# por la media aritmetica de la columna 'cantidad_habitantes' 
df["cantidad_habitantes"].fillna(value=mean_habitantes, inplace=True)
df
```

```python
# creamos un nuevo dataframe a partir de la concatenaci√≥n el dataframe original con otro dataframe secundario.
# ignore_index = True, para ignorar el indice original de los dataframe y iniciar un recuento nuevo para el indice.   
df2 = pd.concat([df,df1], ignore_index=True)
df2
```

```python
# reemplazamos un valor especifico de la columna ciudad por un valor num√©rico
df2["ciudad"] = df2["ciudad"].replace(['santiago'],5)
df2
```

```python
# reemplazamos multiples valores de la columna ciudad  por un espec√≠fico valor num√©rico para cada uno.
df2 = df2.replace({'ciudad':{'concepcion':6,'temuco':7, 'Punta arenas':8 }})
df2
```

```python
# Para jugar con los datos, reemplazamos un valor especifico de la columna "cantidad_habitantes" por un valor nulo, usando np.nan(). 
df2["cantidad_habitantes"] = df2["cantidad_habitantes"].replace([200.0],np.nan)
df2
```

```python
# sumamos la cantidad de datos nulos que aparece en la columna "cantidad_habitantes."
df2["cantidad_habitantes"].isnull().sum()
```
