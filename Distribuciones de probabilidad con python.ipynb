
## **Funcin de Masa de Probabilidad**

Otra forma de representar a las distribuciones discretas es utilizando su Funcin de Masa de Probabilidad o FMP, la cual relaciona cada valor con su probabilidad en lugar de su frecuencia como vimos anteriormente. Esta funcin es normalizada de forma tal que el valor total de probabilidad sea 
- 1. La ventaja que nos ofrece utilizar la FMP es que podemos comparar dos distribuciones sin necesidad de ser confundidos por las diferencias en el tamao de las muestras. Tambin debemos tener en cuenta que FMP funciona bien si el nmero de valores es pequeo; pero a medida que el nmero de valores aumenta, la probabilidad asociada a cada valor se hace cada vez ms pequea y el efecto del ruido aleatorio aumenta. Veamos un ejemplo con Python.


```python
# graficando FMP

# parametros de forma de la distribucin binomial
n = 30 
p = 0.4

# parametros de forma de la distribucin binomial
n_1 = 20
p_1 = 0.3
```

```python
from scipy.stats import binom

x = np.arange(binom.ppf(0.01,n,p), binom.ppf(0.99,n , p))

x_1 = np.arange(binom.ppf(0.01, n_1, p_1), binom.ppf(0.99, n_1, p_1))
```

```python
fmp = binom.pmf(x,n,p) # funcin de masa de probabilidad
fmp_1 = binom.pmf(x_1, n_1, p_1) # funcin de masa de probabilidad
```

```python
plt.plot(x,fmp, '--')
plt.plot(x_1, fmp_1)
plt.vlines(x,0,fmp,colors='b', lw=5, alpha= 0.5)
plt.vlines(x_1,0,fmp_1,colors='g', lw=5, alpha= 0.5)

plt.title('Funcin de masa de probabilidad')
plt.ylabel('probabilidad')
plt.xlabel('valores')
plt.show()
```

Si queremos evitar los problemas que se generan con FMP cuando el nmero de valores es muy grande, podemos recurrir a utilizar la Funcin de Distribucin Acumulada o FDA, para representar a nuestras distribuciones, tanto discretas como continuas. Esta funcin relaciona los valores con su correspondiente percentil; es decir que va a describir la probabilidad de que una variable aleatoria X sujeta a cierta ley de distribucin de probabilidad se site en la zona de valores menores o iguales a x.

```python
fda_binom = binom.cdf(x,n,p) # funcion de distribucion acomulada
```

```python
fda_normal = norm(10, 1.2).cdf(x_1) # funcion de distribucion acomulada
```

## **Distribucin Poisson**

La Distribucin Poisson esta dada por la formula:

# $p(r;\mu) = \frac{\mu^r e^{-\mu}}{r!}$

En dnde **r**
es un entero **(r0)** y **$\mu$** es un nmero real positivo. La Distribucin Poisson describe la probabilidad de encontrar exactamente **r** eventos en un lapso de tiempo si los acontecimientos se producen de forma independiente a una velocidad constante **$\mu$**.

Es una de las distribuciones ms utilizadas en estadstica con varias aplicaciones; como por ejemplo describir el nmero de fallos en un lote de materiales o la cantidad de llegadas por hora a un centro de servicios.

En Python la podemos generar fcilmente con la ayuda de scipy.stats, paquete que utilizaremos para representar a todas las restantes distribuciones a lo largo de todo el artculo.

```python
from scipy.stats import poisson
# graficando Poisson

mu = 3.6 #parametro de forma
poisson = poisson(mu) #distribucion
x = np.arange(poisson.ppf(0.01), poisson.ppf(0.99))
fmp = poisson.pmf(x) # funcion de masa de probabilidad 
```

```python
plt.plot(x,fmp,'--')
plt.vlines(x,0,fmp, colors='b',  lw=55, alpha=0.5)
plt.title('Distribucin Poisson')
plt.ylabel('probabilidad')
plt.xlabel('valores')
plt.show()
```

```python
# histograma 
aleatorios = poisson.rvs(1000) # genera aleatorios

cuenta, cajas, ignorar = plt.hist(aleatorios,20)
plt.ylabel('frecuencia')
plt.xlabel('valores')
plt.title('Histograma Poisson')
plt.show()
```

## **Distribucin Binomial**

La Distribucin Binomial esta dada por la formula:

### $p(r;N,p) = \binom{n}{r} P^r(1-p)^{N-r}$

En dnde **r**
con la condicin 0rN y el parmetro **N** (N>0) son enteros; y el parmetro **p** $(0p1)$ es un nmero real. La Distribucin Binomial describe la probabilidad de exactamente **r** xitos en **N** pruebas si la probabilidad de xito en una sola prueba es **p**.


```python
# graficando binomial

N = 30 # parametros de forma
p = 0.4

binomial = binom(N,p) # Distribucin
x = np.arange(binomial.ppf(0.01), binomial.ppf(0.99))

fmp = binomial.pmf(x) # Funcin de masa de probabilidad

plt.plot(x,fmp, '--')
plt.vlines(x, 0, fmp, colors='b', lw =5, alpha= 0.5)
plt.title('Distribucin Binomial')
plt.ylabel('probabilidad')
plt.xlabel('valores')
plt.show()
```

```python
aleatorios = binomial.rvs(1000) # genera aleatorios

cuenta, cajas, ignorar = plt.hist(aleatorios, 20)

plt.ylabel('frequencia')
plt.xlabel('valores')
plt.title('Histograma Binomial')
plt.show()
```

## **Distribucin Geomtrica**

La Distribucin Geomtrica esta dada por la formula:


$p(r;p)=p(1p)r1$

En dnde r1

La Distribucin Geomtrica expresa la probabilidad de tener que esperar exactamente **r** pruebas hasta encontrar el primer xito si la probabilidad de xito en una sola prueba es **p**. ## **Distribucin Hipergeomtrica**

La Distribucin Hipergeomtrica esta dada por la formula:

## $p(r;n,N,M) = \frac{\binom{M}{r}\binom{N-M}{n-r}}{\binom{N}{n}}$

En dnde el valor de **r**
esta limitado por $max(0,nN+M)$ y $min(n,M)$ inclusive; y los parmetros $n (1nN)$, $N (N1)$ y $M (M1)$ son todos nmeros enteros. La Distribucin Hipergeomtrica describe experimentos en donde se seleccionan los elementos al azar sin reemplazo (se evita seleccionar el mismo elemento ms de una vez). Ms precisamente, supongamos que tenemos $N$ elementos de los cuales $M$ tienen un cierto atributo (y $NM$ no tiene). Si escogemos n elementos al azar sin reemplazo, $p(r)$ es la probabilidad de que exactamente $r$ de los elementos seleccionados provienen del grupo con el atributo.

```python
from scipy.stats import hypergeom
# graficando hipergeomtrica

M,n,N  = 30,10,12 # parametros de forma
hipergeometrica = hypergeom(M,n,N) # Distribucin
x = np.arange(0, n+1)
fmp = hipergeometrica.pmf(x) #funcion de masa de probabilidad
plt.plot(x,fmp,'--')
plt.vlines(x, 0 , fmp, colors='b', lw=5, alpha=0.5)
plt.title('Distribucin hipergeomtrica')
plt.ylabel('probabilidad')
plt.xlabel('valores')
plt.show()
```

```python
#histograma 
aleatorios = hipergeometrica.rvs(1000) #generar aleatorios
cuenta, cajas, ignorar = plt.hist(aleatorios, 20)
plt.ylabel('frecuencia')
plt.xlabel('valores')
plt.title('Histograma hipergeomtrica')
plt.show()
```

## **Distribucin de Normal**

La Distribucin Normal, o tambin llamada Distribucin de Gauss, es aplicable a un amplio rango de problemas, lo que la convierte en la distribucin ms utilizada en estadstica; esta dada por la formula:

$p(x;\mu,\sigma) = \frac{1}{\sigma \sqrt{\pi}} e^{{\frac{-1}{2}}{(\frac{x-\mu}{\sigma})}^2}$

En dnde ***** es el parmetro de ubicacin, y va a ser igual a la media aritmtica y **2** es el desvo estndar. Algunos ejemplos de variables asociadas a fenmenos naturales que siguen el modelo de la Distribucin Normal son:

- caractersticas morfolgicas de individuos, como la estatura;
- caractersticas sociolgicas, como el consumo de cierto producto por un mismo grupo de individuos;
- caractersticas psicolgicas, como el cociente intelectual;
- nivel de ruido en telecomunicaciones;
-errores cometidos al medir ciertas magnitudes;
- etc.


```python
from scipy.stats import norm

# graficando normal
mu = 0 # media
sigma = 0.02  # desvio estandar

normal = norm(mu,sigma)
x = np.linspace(normal.ppf(0.01), normal.ppf(0.99),100)
fp = normal.pdf(x)
plt.plot(x,fp)
plt.title('Distribucion Normal')
plt.ylabel('probabilidad')
plt.xlabel('valores')
plt.show()
```

```python
#histograma 
aleatorios = normal.rvs(1000) # genera aleatorios

cuenta, cajas, ignorar = plt.hist(aleatorios, 20)
plt.ylabel('frecuencia')
plt.xlabel('valores')
plt.title('Histograma Normal')
plt.show()
```

## **Distribucin Gamma**

En dnde los parmetros $a$

# **Distribucin Beta**

La Distribucin Beta esta dada por la formula:

En dnde los parmetros $p$

Las aplicaciones de la Distribucin Beta incluyen el modelado de variables aleatorias que tienen un rango finito de $a$ hasta $b$. Un ejemplo de ello es la distribucin de los tiempos de actividad en las redes de proyectos. La Distribucin Beta se utiliza tambin con frecuencia como una probabilidad a priori para proporciones binomiales en el anlisis bayesiano.

```python
from scipy.stats import beta
# graficando Beta

a,b = 2.3, 0.6 #parametros de forma
beta = beta(a,b)
x = np.linspace(beta.ppf(0.01), beta.ppf(0.99), 100)
fp = beta.pdf(x) #Funcion de probabilidad
plt.plot(x,fp)
plt.title('Distribucin beta')
plt.ylabel('probabilidad')
plt.xlabel('valores')
plt.show()
```

# **Distribucin T de Student**

La Distribucin t de Student esta dada por la funcin:

### $p(t;n) = \frac{\Gamma(\frac{n+1}{2})}{\sqrt n \pi \Gamma (\frac{n}{2})} (1+\frac{t^2}2{})^{-\frac{n+1}{2}}$

En dnde la variable $t$
es un nmero real y el parmetro $n$ es un nmero entero positivo. La Distribucin $t$ de Student es utilizada para probar si la diferencia entre las medias de dos muestras de observaciones es estadsticamente significativa. Por ejemplo, las alturas de una muestra aleatoria de los jugadores de baloncesto podra compararse con las alturas de una muestra aleatoria de jugadores de ftbol; esta distribucin nos podra ayudar a determinar si un grupo es significativamente ms alto que el otro.

```python
from scipy.stats import t
#graficando t de studnet

df = 50 # parametro de forma
t = t(df)
x = np.linspace(t.ppf(0.01), t.ppf(0.99), 100)
plt.plot(x,fp)
plt.title('Distribucin de t de student')
plt.ylabel('probabilidad')
plt.xlabel('valores')
plt.show()
```

# **Distribucin de Pareto**

La Distribucin de Pareto esta dada por la funcin:
# 
$p(x;\alpha,k) = \frac{\alpha k^{\alpha}}{x^{\alpha+1}}$

En dnde la variable $xk$
y el parmetro $>0$ son nmeros reales. Esta distribucin fue introducida por su inventor, Vilfredo Pareto, con el fin de explicar la distribucin de los salarios en la sociedad. La Distribucin de Pareto se describe a menudo como la base de la regla 80/20. Por ejemplo, el 80% de las quejas de los clientes con respecto al funcionamiento de su vehculo por lo general surgen del 20% de los componentes.

```python
from scipy.stats import pareto 
# graficando pareto
k = 2.3 # pareto de forma
pareto = pareto(k)
x = np.linspace(pareto.ppf(0.01),pareto.ppf(0.99), 100)

fp = pareto.pdf(x) # funcion de probabilidad
plt.plot(x,fp)
plt.title('Distribucin de Pareto')
plt.ylabel('probabilidad')
plt.xlabel('valores')
plt.show()
```

```python
#histograma 
aleatorios = pareto.rvs(1000) #genera aleatorios
cuenta, cajas, ignorar = plt.hist(aleatorios, 20)


plt.ylabel('frecuencia')
plt.xlabel('valores')
plt.title('Histograma de Pareto')
plt.show()
```
